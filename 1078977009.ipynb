# Reflectance and Absorbance Spectra

Reflectance spectra for binding (for image see [3D Model](https://skfb.ly/ouTVr))

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()

# prepare some data
x1 = [460]
x2 = [550]
x3 = [640]
y1 = [0.03]
y2 = [0.05]
y3 = [0.10]

# create a new plot with a title and axis labels
p = figure(title="History of the Catholic Church in Australasia (Binding) 2022-06-06", 
           x_axis_label="WAVELENGTH, nm", y_axis_label="% REFLECTANCE", width=600, height=600) 

# add multiple renderers
p.diamond(x=x1, y=y1, size=15, color="Blue", line_width=2)
p.diamond(x=x2, y=y2, size=15, color="Green", line_width=2)
p.diamond(x=x3, y=y3, size=15, color="Red", line_width=2)

# show the results
show(p)
```
