### **Condition report with reflectance and absorbance datasets**

**OCLC No.: 2756214**

Wight, J, Andrew Agnew, George Cruikshank, Effingham Wilson, Thomas Hurst, John Thompson, Samuel Williams, et al. *Sunday in London*. London: Effingham Wilson, 1833.

2022-05-06

**Binding**

Bound in late nineteenth century brown goatskin with gold tooling on cover.

[Condition review as 3D Scan](https://skfb.ly/ousUt)

**Reflectance of Binding**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()

# prepare some data
x1 = [460]
x2 = [550]
x3 = [640]
y1 = [2]
y2 = [4]
y3 = [7]

# create a new plot with a title and axis labels
p = figure(title="Reflectance spectra for Sunday in London (Binding) 2022-05-06", 
           x_axis_label="WAVELENGTH, nm", y_axis_label="% REFLECTANCE", width=600, height=600)

# add multiple renderers
p.diamond(x=x1, y=y1, size=15, color="Blue", line_width=2)
p.diamond(x=x2, y=y2, size=15, color="Green", line_width=2)
p.diamond(x=x3, y=y3, size=15, color="Red", line_width=2)

# show the results
show(p)
```

```python
from bokeh.io import output_file, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.transform import dodge

output_file("dodged_bars.html")

fruits = ['Aged 189 Years', 'Aged 190 Years']
years = ['L', 'a', 'b']

data = {'fruits' : fruits,
        'L'   : [90.79, 0],
        'a'   : [1.78, 0],
        'b'   : [11.83, 0]}

source = ColumnDataSource(data=data)

p = figure(x_range=fruits, y_range=(-10, 127), height=800, title="CIE LAB Tristimulus Values for Title Page",
           toolbar_location=None, tools="")

p.vbar(x=dodge('fruits', -0.25, range=p.x_range), top='L', width=0.2, source=source,
       color="#c9d9d3", legend_label="L")

p.vbar(x=dodge('fruits',  0.0,  range=p.x_range), top='a', width=0.2, source=source,
       color="#718dbf", legend_label="a")

p.vbar(x=dodge('fruits',  0.25, range=p.x_range), top='b', width=0.2, source=source,
       color="#e84d60", legend_label="b")

p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
